#    Bank Loan Risk Analysis  #


```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("credit_test.csv")
df
```

### Q1. Find the average current loan amount by home ownership type.

```python
df.groupby('Home Ownership')['Current Loan Amount'].mean()
```

### Q2. Calculate the average credit score for each home ownership type.

```python
df.groupby('Home Ownership')['Credit Score'].mean()
```

### Q-3. Identify customers with credit score below 500.

```python
low_credit = df[df['Credit Score'] < 500]
print(low_credit)
```

### Q4. Analyse loan amount distribution based on loan term (short-term vs long-term).

```python
df.groupby('Term')['Current Loan Amount'].describe()
```

### Q5. Calculate the average annual income of customers.

```python
df['Annual Income'].mean()
```

### Q6. Count the number of customers in each home ownership category.

```python
df['Home Ownership'].value_counts()
```

### Q7. Identify customers whose current loan amount is greater than their annual income.

```python
high_loan = df[df['Current Loan Amount'] > df['Annual Income']]
print(high_loan)
```

### Q8. Find the top 10 customers with the highest current loan amount.

```python
top10 = df.sort_values(by='Current Loan Amount', ascending=False).head(10)
print(top10)
```

### Q9. Calculate the average credit score based on the loan term.

```python
df.groupby('Term')['Credit Score'].mean()
```

### Q10. Identify high-risk customers based on:
### • Low credit score
### • High monthly debt

```python
high_risk = df[(df['Credit Score'] < 500) & (df["Monthly Debt"] > df['Monthly Debt'].mean()) ]
print(high_risk)
```
